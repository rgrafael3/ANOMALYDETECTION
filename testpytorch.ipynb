```python
# -*- coding: utf-8 -*-
"""
Created on Wed May 22 23:56:15 2024

@author: rgraf
"""




import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split








#Check accuracy of the model
def checkTestData(model, testing_dataXY):
    # compute accuracy (no_grad is optional)

    Xt=testing_dataXY[:,0:8]
    Yt=testing_dataXY[:,8]

    Xt = torch.tensor(Xt, dtype=torch.float32)
    Yt = torch.tensor(Yt, dtype=torch.float32).reshape(-1, 1)

    with torch.no_grad():
        y_pred = model(Xt)
    accuracy = (y_pred.round() == Yt).float().mean()
    return accuracy

## TORCH
dtype = torch.FloatTensor

# load the dataset, split into input (X) and output (y) variables
dataset = np.loadtxt('pima-indians-diabetes.data.csv', delimiter=',')
training_dataXY, testing_dataXY = train_test_split(dataset, test_size=0.2, random_state=25)

X = training_dataXY[:,0:8]
y = training_dataXY[:,8]


X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32).reshape(-1, 1)

model = nn.Sequential(
    nn.Linear(8, 12),
    nn.ReLU(),
    #nn.Sigmoid(),
    nn.Linear(12, 8),
    #nn.ReLU(),
    nn.ReLU(),
    nn.Linear(8, 1),
    nn.Sigmoid()
)

# train the model
loss_fn   = nn.BCELoss()  # binary cross entropy
optimizer = optim.Adam(model.parameters(), lr=0.001)

n_epochs = 15000
batch_size = 1000
lastacc=0


for epoch in range(n_epochs):

    permutation = torch.randperm(X.size()[0])
    for i in range(0, len(X), batch_size):
        indices = permutation[i:i+batch_size]

        Xbatch = X[indices]
        y_pred = model(Xbatch)
        ybatch = y[indices]


        loss = loss_fn(y_pred, ybatch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f'Finished epoch {epoch}, latest loss {loss}')
    newacc=checkTestData(model, testing_dataXY)
    print ('training acc',checkTestData(model, training_dataXY))
    print (newacc)
    print
    if lastacc > newacc:
        if epoch >10000:
            break
    else:
        lastacc=newacc
# compute accuracy (no_grad is optional)

Xt=testing_dataXY[:,0:8]
Yt=testing_dataXY[:,8]

Xt = torch.tensor(Xt, dtype=torch.float32)
Yt = torch.tensor(Yt, dtype=torch.float32).reshape(-1, 1)

with torch.no_grad():
    y_pred = model(Xt)
accuracy = (y_pred.round() == Yt).float().mean()
print(f"Accuracy {accuracy}")
```
